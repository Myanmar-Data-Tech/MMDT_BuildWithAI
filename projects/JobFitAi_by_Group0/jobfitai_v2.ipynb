```python
# --- 1. SETUP & INSTALLATION ---
# Install the Google Generative AI SDK
!pip install -U -q google-generativeai

import os
import base64
import google.generativeai as genai
from getpass import getpass
from IPython.display import display, Markdown

# --- 2. AUTHENTICATION ---
# Securely handle your API key
api_key = getpass('Enter your Gemini API Key: ')
genai.configure(api_key=api_key)
```

```python
# --- 3. THE MINIMALIST LOGIC ---
def process_cv(pdf_path, job_description):
    """
    Processes a PDF CV against a JD using a minimalist recruiter persona.
    Logic mirrors the provided TypeScript implementation.
    """

    # Encode PDF to Base64 to pass as inline data
    with open(pdf_path, "rb") as f:
        pdf_base64 = base64.b64encode(f.read()).decode("utf-8")

    # Strict System Instruction to enforce the minimalist UI/UX
    system_instruction = """
    You are a Minimalist Technical Recruiter. Your task is to process a CV (PDF) against a provided Job Description.


    STRICT OUTPUT RULES:
    - Use bullet points only for lists.
    - No introductory or concluding remarks.
    - Be blunt and objective.
    - Follow the exact REPORT STRUCTURE provided below.

    REPORT STRUCTURE:
    ### 👤 Candidate: [Name]
    - **Experience:** [Total years]
    - **Top Skills:** [List top 5]

    ### ⚖️ Alignment
    - **Match Score:** [X/100]
    - **Verdict:** [Strong Fit / Promising Candidate / Needs Development]

    ### 🌱 Strengths
    - [List key strengths relevant to the role]

    ### 🚀 To-Do
    - [3 short career improvement tips]
    - [1 recommended online course with platform name]
    - [1 suggested certification if relevant]
    """

    # Initialize the model (Gemini 1.5 Flash is recommended for speed/cost)
    model = genai.GenerativeModel(
        model_name="gemini-2.5-flash",
        system_instruction=system_instruction
    )

    # Generate Content
    response = model.generate_content([
        {
            "mime_type": "application/pdf",
            "data": pdf_base64
        },
        f"Job Description:\n{job_description}"
    ], generation_config={"temperature": 0.1})

    return response.text
```

```python
# --- 4. EXECUTION ---
# /alex_CV.pdf
# /jordan_CV.pdf
```

```python
import os

while True:
    print("\n--- CV Analysis Tool (Type 'q' to exit) ---")

    cv_filename = input('Enter your file name (e.g., jordan_CV.pdf): ').strip()

    # Check if the user wants to exit
    if cv_filename.lower() == 'q':
        print("Exiting tool. Good luck with your applications!")
        break

    jd_text = input('Enter the JD of the position you want to apply: ').strip()
    if jd_text.lower() == 'q':
        break

    # Validation and Processing
    if os.path.exists(cv_filename):
        print(f"⏳ Analyzing {cv_filename}...")

        # Calling your processing function
        report = process_cv(cv_filename, jd_text)

        display(Markdown("---"))
        display(Markdown(report))

        # Optional: Ask if they want to do another after a success
        continue_choice = input("\nWould you like to analyze another? (y/n): ").lower()
        if continue_choice != 'y':
            break
    else:
        print(f"❌ Error: {cv_filename} not found.")
        print("Please ensure the file is uploaded to the correct directory and try again.")
```
